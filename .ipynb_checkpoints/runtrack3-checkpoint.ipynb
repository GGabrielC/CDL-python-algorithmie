```python
import re
import matplotlib.pyplot as plt
   
with open("data.txt", "r") as f:
    words_pattern = '[a-z]+'
    words = re.findall(words_pattern, f.read(), flags=re.IGNORECASE)
    letters = "".join(words)
    letters = letters.lower()
    alphabet = set(letters)
    d={ l:{i:0 for i in alphabet} for l in alphabet }
    [d[l].update( {nl: 1+d[l][nl]} ) for (l, nl) in zip(letters[:-1], letters[1:])]
    #d={ l:[]  for l in set(letters) }
    #[d[l].append(nl) for (l, nl) in zip(letters[:-1], letters[1:])]
```

```python
fig, ax = plt.subplots()
for k, v in d.items():
    points = [(x,y) for (x,y) in v.items()]
    points = sorted(points, key=lambda p: p[0])
    xs, ys = list(zip(*points))
    ys = [i/sum(ys) for i in ys]
    ax.plot(xs, ys, label=k)

ax.legend()
```
